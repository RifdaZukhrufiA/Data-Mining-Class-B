## Tugas Integrasi Data Iklim Harian Surabaya

```python
import pandas as pd
import numpy as np
```

```python
#import data
iklim1 = pd.read_csv('iklim1.csv')
iklim2 = pd.read_csv('iklim2.csv')
```

```python
iklim1
```

```python
iklim2
```

```python
#merge data berdasarkan RR
dataa = pd.merge(iklim1, iklim2, on='Tanggal')
```

```python
dataa
```

```python
#menghapus kolom
data1 = dataa.drop(['Tavg'], axis=1)
data1
```

```python
#menghapus baris
data2 = dataa.drop([2])
data2
```

```python
#menghapus 5 baris pertama
data3 = dataa.drop([0,1,2,3,4])
data3
```

```python
#mengubah urutan kolom
urut1 = iklim1[['Tanggal','RR','Tn','Tx','ff_x','ddd_car','ddd_x']]
urut1
```

```python
#mengubah urutan kolom
urut2 = iklim2[['Tanggal','RR','Tavg','ff_avg','ss','RH_avg']]
urut2
```

```python
#multiplekey
multkey = pd.merge(urut1,urut2,on=['Tanggal','RR'])
multkey
```

```python
#merge methods (inner)
result_left = pd.merge(urut1,urut2,how='left',on=['Tanggal','RR'])
result_left
```

```python
#merge methods (inner)
result_right = pd.merge(urut1,urut2,how='right',on=['Tanggal','RR'])
result_right
```

```python
result_outer = pd.merge(urut1, urut2, how='outer', on=['Tanggal','RR'])
result_outer
```
