```python
#imputasi missing value data numerik
data['Grams_day'] = data['Grams_day'].fillna((data['Grams_day'].mean())) 
data['Packs_year'] = data['Packs_year'].fillna((data['Packs_year'].median()))
data['INR'] = data['INR'].fillna((data['INR'].mean()))
data.isnull().any()
```

```python
#imputasi data kategorik dengan modus
data['Symptoms'] = data.fillna(data['Symptoms'].value_counts().index[0])
data['Alcohol'] = data.fillna(data['Alcohol'].value_counts().index[0])
data['HBsAg'] = data.fillna(data['HBsAg'].value_counts().index[0])
data.isnull().any()
```

```python
#imputasi data kategorik dengan modus sesuai kolom masing2
data = data.apply(lambda x:x.fillna(x.value_counts().index[0]))
data.isnull().any()
```

## Encode

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Encode untuk Gender, Symptoms, Alcohol, Grams_day, Packs_year, dan INR tanpa membuat kolom baru
data['Gender'] = LabelEncoder().fit_transform(data['Gender'])
data['Symptoms'] = LabelEncoder().fit_transform(data['Symptoms'])
data['Alcohol'] = LabelEncoder().fit_transform(data['Alcohol'])
data['Grams_day'] = LabelEncoder().fit_transform(data['Grams_day'])
data['Packs_year'] = LabelEncoder().fit_transform(data['Packs_year'])
data['INR'] = LabelEncoder().fit_transform(data['INR'])
data.head()
```

## Standardize

```python
from sklearn.preprocessing import StandardScaler
```

```python
data['Gender'] = StandardScaler().fit(data[['Gender']]).transform(data[['Gender']])
data['Age'] = StandardScaler().fit(data[['Age']]).transform(data[['Age']])
data['Alcohol'] = StandardScaler().fit(data[['Alcohol']]).transform(data[['Alcohol']])
data['Grams_day'] = StandardScaler().fit(data[['Grams_day']]).transform(data[['Grams_day']])
data['Packs_year'] = StandardScaler().fit(data[['Packs_year']]).transform(data[['Packs_year']])
```

## Normalize

```python
from sklearn.preprocessing import Normalizer 
```

```python
data['Gender'] = Normalizer().fit(data[['Gender']]).transform(data[['Gender']])
data['Age'] = Normalizer().fit(data[['Age']]).transform(data[['Age']])
data['Alcohol'] = Normalizer().fit(data[['Alcohol']]).transform(data[['Alcohol']])
data['Grams_day'] = Normalizer().fit(data[['Grams_day']]).transform(data[['Grams_day']])
data['Packs_year'] = Normalizer().fit(data[['Packs_year']]).transform(data[['Packs_year']])
```
